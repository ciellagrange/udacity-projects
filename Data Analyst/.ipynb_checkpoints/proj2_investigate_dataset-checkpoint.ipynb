```python
movies = pd.read_csv('~/Downloads/tmdb-movies.csv')
movies.head()
```

```python
cast_list = movies['cast'].str.split('|')
```

```python
cast_list.head()
```

```python
dic = {}
for ind, cast in cast_list.items():
    if isinstance(cast, list):
        for c in cast:
            #print(ind, c)
            if c in dic.keys():
                dic[c] += 1
            else:
                dic[c] = 1
```

```python
from operator import itemgetter
sorted(dic.items(), key=itemgetter(1),reverse=True)[:20]
```

```python
company_list = movies['production_companies'].str.split('|')
```

```python
dic2 = {}
for ind, comp in company_list.items():
    if isinstance(comp, list):
        for c in comp:
            #print(ind, c)
            if c in dic2.keys():
                dic2[c] += 1
            else:
                dic2[c] = 1
```

```python
sorted(dic2.items(), key=itemgetter(1),reverse=True)[:20]
```
